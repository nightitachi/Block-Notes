```python
def processed_data(df: pd.DataFrame, is_test=False):
    col_train = ['date', 'store_nbr', 'onpromotion', 'family', 'id', 'sales']
    col_test = ['date', 'family', 'store_nbr', 'onpromotion', 'id']
    col_holidays = ['date', 'type', 'transferred']

    if is_test:
        col_df = col_test
    else:
        col_df = col_train

    # Ensure 'date' columns are of datetime type
    df['date'] = pd.to_datetime(df['date'])
    oil['date'] = pd.to_datetime(oil['date'])
    holidays['date'] = pd.to_datetime(holidays['date'])

    # Merging data
    df = df[col_df].merge(oil, 'left', 'date').merge(holidays[col_holidays], 'left', 'date').rename(
        columns={'type': 'day_type', 'dcoilwtico': 'crude_price'})

    df['day_type'] = df['day_type'].fillna('Work Day')
    df['transferred'] = df['transferred'].fillna(False)

    df['is_holiday'] = np.where(
        (df['day_type'] == 'Work Day'), 0,
        np.where(
            (df['day_type'].isin(['Holiday', 'Additional', 'Event', 'Transfer', 'Bridge'])) &
            (df['transferred'] == False),
            1,
            0
        )
    )

    # Fixing onpromotion column values
    df['onpromotion'] = np.where(df['onpromotion'] == 0.0, 0, 1)

    df['day_number'] = pd.factorize(df['date'])[0] + 1
    df['day_type'] = df.drop(['transferred', 'day_type', 'date'], axis=1)
    df['crude_price'] = df['crude_price'].fillna(method='ffill')

    # Encoding the 'family' column
    family_encoding = pd.get_dummies(df['family'], prefix='family').astype(int)
    df = pd.concat([df, family_encoding], axis=1)
    df = df.drop(['family'], axis=1)

    # Drop missing values
    df = df.dropna()

    return df

df_train = processed_data(train)
df_test = processed_data(test, is_test=True)
```
